```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Etapa 1 - Extract  
Nessa etapa é realizada a extração dos dados. Foi utilizada a biblioteca pandas para leitura do e manipulação do arquivo, sendo possível visualizar as informações em formato de coluna e identificar os tipos de dados de cada coluna e se há dados nulos. 

```python
df = pd.read_csv ('vendas.csv')

df.head()
```

```python
df.info()
```

# Etapa 2 - Transform  
Nessa etapa foi realizada a conversão do tipo de dado presente na coluna 'data_venda' de **object** para **datetime**, além de tratar os valores nulos.

## Conversão do tipo da coluna

```python
df['data_venda'] = pd.to_datetime(df['data_venda'], format='%d/%m/%Y')
print(df.dtypes)
df.head()
```

## Tratar valores nulos

#### Coluna Valor

```python
df.describe()
```

```python
sns.set_style('whitegrid')

plt.figure(figsize=(8,6))
sns.boxplot(data=df, y='valor', color = 'skyblue')
plt.title('Boxplot da coluna Valor')
plt.show()
```

```python
mediana_valor = df['valor'].median()
df.fillna({'valor' : mediana_valor}, inplace=True)
```

```python
df['valor'].isnull().sum()
```

#### Coluna Desconto

```python
df.fillna({'desconto' : 0}, inplace=True)
df['desconto'].isnull().sum()
```

#### Coluna Categoria

```python
df.fillna({'categoria' : "Outros"}, inplace=True)
df['categoria'].isnull().sum()
```

## Adicionar coluna valor_final

```python
df['valor_final'] = df['valor'] - (df['valor'] * df['desconto'] / 100)
print(df['valor_final'])
```

## Adicionar coluna ticket_alto

```python
df['ticket_alto'] = df['valor_final'].apply(lambda x: 'Sim' if x > 1000 else 'Não')
```

```python
print(df['ticket_alto'])
```

# Etapa 3 -Load

```python
df.to_csv('vendas_ETL.csv', index=False)
```
